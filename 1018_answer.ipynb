```python
ex_y_num, ex_x_num = map(int,input().split())

board_lst=list() #가져온 보드

for i in range(ex_y_num):
  board_lst.append(input())

chest_lst=[] # 8*8로 자를 체스판
my_color = [] #칠해야 하는 칸의 개수
num = 0 #my_color의 요소


for o in range(ex_x_num-7): #가로축
  for p in range(ex_y_num-7): #세로축
    for k in range(p,p+8):
      chest_lst.append(board_lst[k][o:o+8])
    #체스판 완성
    first_w = 0
    first_b = 0
    for y in range(8): #세로축
      for x in range(8): #가로축
        if (y + x) % 2 == 0: # 짝수,짝수/홀수,홀수
          if chest_lst[y][x] == "B":
            first_w+=1
          else:
            first_b+=1
        else: #짝수, 홀수/홀수, 짝수
          if chest_lst[y][x] == "W":
            first_w+=1
          else:
            first_b+=1
        
    my_color.append(first_w)
    my_color.append(first_b)
 
    chest_lst.clear()

print(min(my_color))
```
