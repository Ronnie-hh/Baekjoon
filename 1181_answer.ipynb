```python
words=[]
N = int(input())

for n in range(N):
  words.append(input())

words.sort() #알파벳 순으로 정렬
words.sort(key=len) #요소 길이 순으로 정렬

ans =[]

for n in range(N):
  ans.append(words[n])
  if n==N-1:
    break
  if words[n]==words[n+1]:
    ans.remove(words[n+1])

for i in range(len(ans)):
  print(ans[i])

```
